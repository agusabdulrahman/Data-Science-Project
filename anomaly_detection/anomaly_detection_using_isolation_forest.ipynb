```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
file_path = './dataset/temperature.csv'
selected_columns = ['Date', 'Value']
df = pd.read_csv(file_path)
df = df[selected_columns]
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')
```

```python
feature_column = "Value"
contamination=0.010

model = IsolationForest(contamination=contamination)
df['anomalies'] = model.fit_predict(df[feature_column].values.reshape(-1, 1))
```

```python
df.head()
```

```python
# Plot data
plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Value"], label="Temperature", color="blue", alpha=0.7)

# Add anomly
anomalies = df[df["anomalies"] == -1]
plt.scatter(anomalies["Date"], anomalies["Value"], color="red", label="Anomaly", s=50)

plt.title(f"Anomaly Detection using Isolation Forest with contamination: {contamination}", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Temperature", fontsize=12)
plt.legend()
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```
